```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import time
import datetime
```

```python
# Cities data file
csv_file = "Resources/cities.csv"

# Read the data
cities_df = pd.read_csv(csv_file)
cities_df.head()
```

```python
#Date conversion
date =datetime.datetime.fromtimestamp(
        int(cities_df['Date'][0])
    ).strftime('%m/%d/%Y')
date
```

```python
# Generate a scatter plot of Latitude vs Max Temp
x_values = cities_df["Lat"]
y_values = cities_df["Max Temp"]

plt.scatter(x_values,y_values, marker = "o", facecolors = "green", edgecolors = "black", 
            alpha = 0.80)

# Set title and labels
plt.xlabel('Latitude', fontweight='bold')
plt.ylabel('Max Temperature (F)', fontweight='bold')
plt.title("City Latitude vs Max Temperature (F) ({})".format(date))
plt.xlim(-80,100)
plt.ylim(-100, 150)
plt.grid()

# Save image
plt.savefig("./images/lat vs max temp")
```

```python
# Generate a scatter plot of latitude vs Humidity
x_values = cities_df["Lat"]
y_values = cities_df["Humidity"]

plt.scatter(x_values,y_values, marker = "8", facecolors = "red", edgecolors = "black", 
            alpha = 0.80)

# Set title and labels
plt.xlabel('Latitude', fontweight='bold')
plt.ylabel('Humidity (%)', fontweight='bold')
plt.title("City Latitude vs Humidity ({})".format(date))
plt.xlim(-80,100)
plt.ylim(-20, 120)
plt.grid()

# Save image
plt.savefig("./images/lat vs hum")
```

```python
# Generate a scatter plot of latitude vs Cloudiness
x_values = cities_df["Lat"]
y_values = cities_df["Cloudiness"]

plt.scatter(x_values,y_values, marker = "v", facecolors = "blue", edgecolors = "black", 
            alpha = 0.80)

# Set title and labels
plt.xlabel('Latitude', fontweight='bold')
plt.ylabel('Cloudiness (%)', fontweight='bold')
plt.title("City Latitude vs Cloudiness ({})".format(date))
plt.xlim(-80,100)
plt.ylim(-20, 120)
plt.grid()

# Save image
plt.savefig("./images/lat vs cloud")
```

```python
# Generate a scatter plot of latitude vs Wind Speed (mph)
x_values = cities_df["Lat"]
y_values = cities_df["Wind Speed"]

plt.scatter(x_values,y_values, marker = "d", facecolors = "yellow", edgecolors = "black", 
            alpha = 0.80)

# Set title and labels
plt.xlabel('Latitude', fontweight='bold')
plt.ylabel('Wind Speed (mph)', fontweight='bold')
plt.title("City Latitude vs Wind Speed ({})".format(date))
plt.xlim(-80,100)
plt.ylim(-5, 40)
plt.grid()

# Save image
plt.savefig("./images/lat vs wind speed")
```
